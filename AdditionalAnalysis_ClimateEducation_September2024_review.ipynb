```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(12, 10))
mask = np.triu(np.ones_like(df3))
sns.heatmap(df3, ax=ax, annot=True, cmap='YlGnBu',fmt=".2f",linewidth=1,mask = mask)
ax.set_xticklabels(labels =ax.get_xticklabels(), rotation=90, ha='right', rotation_mode="anchor")
ax.tick_params(axis='both', labelsize=10)
plt.tight_layout()
plt.savefig('../../Data/ClimateEducation/Figures_092024/Heatmap_topics_all.eps',dpi='figure',
        bbox_inches='tight', pad_inches=0.2,
        facecolor='auto', edgecolor='auto',
        backend=None )
```
